```python
#import libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Data Cleaning

#creating a DataFrame and dowloading data
df = pd.read_csv("../data/diabetic_readmission_data.csv") 

#dropping weight (all '?'), payer_code (irrelevant), medical_specialty (almost all'?'), patient_nbr (irrelevant), encounter_id (irrelevant)
df.drop(['weight','payer_code','medical_specialty', 'patient_nbr', 'encounter_id'], axis=1, inplace = True)

#display all columns in jupyter notebook
pd.options.display.max_columns = None

#replace all instances of grouped age with the last value in the group.
df['age'].replace(to_replace = '[0-10)', value = 10, inplace=True)
df['age'].replace(to_replace = '[10-20)', value = 20, inplace=True)
df['age'].replace(to_replace = '[20-30)', value = 30, inplace=True)
df['age'].replace(to_replace = '[30-40)', value = 40, inplace=True)
df['age'].replace(to_replace = '[40-50)', value = 50, inplace=True)
df['age'].replace(to_replace = '[50-60)', value = 60, inplace=True)
df['age'].replace(to_replace = '[60-70)', value = 70, inplace=True)
df['age'].replace(to_replace = '[70-80)', value = 80, inplace=True)
df['age'].replace(to_replace = '[80-90)', value = 90, inplace=True)
df['age'].replace(to_replace = '[90-100)', value = 100, inplace=True)

#replace M and F with 1 and 0 respectively
df['gender'].replace(to_replace ='Female', value= 0, inplace=True)
df['gender'].replace(to_replace ='Male', value= 1, inplace=True)

#remove all race values with '?'
df1 = df[df.race != '?']

df1
```

```python
#Data Investigation

#creat pivot table grouped by time in hospital 
df1.groupby(['race','gender'])['readmit'].sum()
```

```python
df1.groupby(['readmitted','time_in_hospital']).mean()
```

```python
#Feature Engineering

#create a column to group <30 and >30 into True and 'NO' to False.
df1['readmit'] = df1['readmitted'].map({'NO': False, '>30': True, '<30': True})
df1
```
